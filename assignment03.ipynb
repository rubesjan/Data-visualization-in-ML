p≈ôeveden√≠ sloupc≈Ø typu object na ƒç√≠selnou reprezentaci:

ExerciseAngina obsahuje bin√°rn√≠ data z popisu na Kaggle

ST_Slope obsahuje oridn√°ln√≠ data, tak≈æe ho p≈ôevedu na ƒç√≠selnou reprezentaci

```python
# train dataset
xtrain["ExerciseAngina"].replace({"Y":1, "N":0}, inplace=True)

slope_category = pd.api.types.CategoricalDtype(categories=["Up", "Flat", "Down"], ordered=True)
xtrain["ST_Slope"] = xtrain["ST_Slope"].astype(slope_category)
xtrain["ST_Slope"] = xtrain["ST_Slope"].cat.codes

# test dataset
xtest["ExerciseAngina"].replace({"Y":1, "N":0}, inplace=True)

slope_category = pd.api.types.CategoricalDtype(categories=["Up", "Flat", "Down"], ordered=True)
xtest["ST_Slope"] = xtest["ST_Slope"].astype(slope_category)
xtest["ST_Slope"] = xtest["ST_Slope"].cat.codes
```

p≈ôeveden√≠ zbyl√Ωch object sloupc≈Ø na ƒç√≠selnou reprezentaci pomoc√≠ one hot encoding:

```python
# train dataset
xtrain["Sex"] = xtrain["Sex"].replace({"M": 1, "F":0})
xtrain.rename({"Sex": "Male"}, inplace=True)

xtrain = pd.get_dummies(xtrain, columns=xtrain.select_dtypes(include='object').columns)
xtrain.replace({False: 0, True: 1}, inplace=True)

# test dataset
xtest["Sex"] = xtest["Sex"].replace({"M": 1, "F":0})
xtest.rename({"Sex": "Male"}, inplace=True)

xtest = pd.get_dummies(xtest, columns=xtest.select_dtypes(include='object').columns)
xtest.replace({False: 0, True: 1}, inplace=True)

data = xtrain
display(data.info())
print(f"data shape: {data.shape}")
```

rozdƒõlen√≠ tr√©novac√≠ch dat na validaƒçn√≠ a tr√©novac√≠:

```python
xtrain, xval, ytrain, yval = train_test_split(xtrain, ytrain, test_size=0.25, random_state=random_seed)
```

# Tr√©nov√°n√≠ modelu

Vybral jsem si model rozhodovac√≠ho stromu. Mysl√≠m si, ≈æe je pro tuto √∫lohu je vhodn√Ω a d√° se i vizualizovat.
Budu hledat nejlep≈°√≠ hodnotu tƒõchto hyperparametr≈Ø:

    max_depth - <3, 15) - maxim√°ln√≠ hloubka stromu, ƒç√≠m je vƒõt≈°√≠, t√≠m v√≠ce  se strom m≈Ø≈æe p≈ôeuƒçit
    min_samples_leaf - <2, 10) - minim√°ln√≠ poƒçet z√°znam≈Ø v ka≈æd√©m listu
    criterion - {gini, entropy} - krit√©rium na hodnocen√≠ neuspo≈ô√°danosti v mno≈æinƒõ (vysvƒõtlovan√© promƒõnn√©)


jako ztr√°tovou funkci pou≈æiji binary cross-entropy loss:
    
    L(y,p)= -y‚ãÖlog(p) - (1-y)‚ãÖlog(1‚àíp)

hled√°n√≠ nejlep≈°√≠ch hyperparamtr≈Ø modelu:

```python
hyperparams = {
    "max_depth": range(3, 15),
    "min_samples_leaf": range(2, 10),
    "criterion": ["gini", "entropy"]
}

val_loss = []

hyperparams = ParameterGrid(hyperparams)

for param in hyperparams:
    tree = DecisionTreeClassifier(**param, random_state=random_seed)
    tree.fit(xtrain, ytrain)
    ypred = tree.predict_proba(xval)
    val_loss.append(metrics.log_loss(yval, ypred))
```

vyps√°n√≠ nejni≈æ≈°√≠ cross-entropy a nejlep≈°√≠ch hyperparametr≈Ø:

```python
def get_argmin(l):
    m = np.min(l)
    res = []
    for i in range(len(l)):
        if m == l[i]:
            res.append(i)
    return res

best_params = get_argmin(val_loss)
best_params = [hyperparams[i] for i in best_params]

print(f"Minim√°ln√≠ cross-entropy loss: {round(np.min(val_loss), 4)}")
print("\nNejlep≈°√≠ hyperparametry:")
for param in best_params:
    print(param)

val_loss = np.min(val_loss)
```

zde je vidƒõt, ≈æe ztr√°tov√° funkce nab√Ωv√° nejmen≈°√≠ hodnoty pro tyto hyperparametry 

```python
best_params = best_params[-1]
```

```python
# preparation for graph

val_gini_loss = []
val_entropy_loss = []

for leaf in range(1, 10):
    tree = DecisionTreeClassifier(max_depth=3, criterion="gini", min_samples_leaf=leaf, random_state=random_seed)
    tree.fit(xtrain, ytrain)
    val_gini_loss.append(metrics.log_loss(yval, tree.predict_proba(xval)))
    
    tree = DecisionTreeClassifier(max_depth=3, criterion="entropy", min_samples_leaf=leaf, random_state=random_seed)
    tree.fit(xtrain, ytrain)
    val_entropy_loss.append(metrics.log_loss(yval, tree.predict_proba(xval)))
```
